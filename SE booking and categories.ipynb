```python
def Booking():

    print("******BOOK YOUR ROOM FOR YOUR MEMORABLE STAY******")

    checkin_date=str(input("Enter check-in date: "))
    checkin.append(checkin_date)
    checkin_date=checkin_date.split('/')
    date_segments=checkin_date
    date_segments[0]=int(date_segments[0])
    date_segments[1]=int(date_segments[1])
    date_segments[2]=int(date_segments[2])

    checkout_date=str(input("Enter check-out date: "))
    checkout.append(checkout_date)
    checkout_date=checkout_date.split('/')
    date_out_seg=checkout_date
    date_out_seg[0]=int(date_out_seg[0])
    date_out_seg[1]=int(date_out_seg[1])
    date_out_seg[2]=int(date_out_seg[2])


    d1 = datetime.datetime(date_out_seg[2],date_out_seg[1],date_out_seg[0])
    d2 = datetime.datetime(date_out_seg[2],date_out_seg[1],date_out_seg[0])
    stay = (d2-d1).days
    days.append(stay)

    print("******AVAILABLE CATEGORIES OF ROOMS******")
    print(" 1. STANDARD")
    print(" 2. TWIN BED")
    print(" 3. MINI SUITE")
    print(" 4. LUXURY")
    print("******PRESS 0 TO EXPLORE ROOM INFORMATION******")
    choice=int(input("->"))

    # if-conditions to display allotted room
    # type and it's price
    if choice==0:
        Rooms_Details()
        print("******PRESS 0 TO GO BACK TO BOOKING PAGE******")
        choice=int(input("->"))
        if choice==0:
            print("******AVAILABLE CATEGORIES OF ROOMS******")
            print(" 1. STANDARD")
            print(" 2. TWIN BED")
            print(" 3. MINI SUITE")
            print(" 4. LUXURY")
            print(("******CHOOSE YOUR DESIRED CATEGORY FOR BOOKING******"))
            choice=int(input("->"))


            if choice==1:
                r=[]
                room.append('STANDARD')
                print("Category: standard")
        #       generating random room numbers
                for i in range(1,10):
                    r.append(random.randrange(40)+300)

                print("Available rooms in this category: ",r)
                room_no=r[5]
                price.append(2000)
                print("Charges: 2000 per night")
                
            elif choice==2:
                r=[]
                room.append('TWIN BED')
                print("Category: Twin Bed") 
                for i in range(1,10):
                    r.append(random.randrange(40)+300)
                print("Available rooms in this category: ",r)
                room_no=r[5]
                price.append(3000)
                print("Charges: 3000 per night")
                
            elif choice==3:
                r=[]
                room.append('MINI SUITE')
                print("Category: Mini Suite")
                for i in range(1,10):
                    r.append(random.randrange(40)+300)
                print("Available rooms in this category: ",r)
                room_no=r[5]
                price.append(4000)
                print("Charges: 4000 per night")
            elif choice==4:
                r=[]
                room.append('LUXURY')
                print("Category: Luxury")
                for i in range(1,10):
                    r.append(random.randrange(40)+300)          
                print("Available rooms in this category: ",r)
                room_no=r[5]
                price.append(5000)
                print("Charges: 5000 per night")
            else:
                print("INVALID CATEGORY")


    # alloting random id for customer
    cust_id = random.randrange(40)+10


    # checks if allotted room no. & customer
    # id already not allotted
    while room_no in room_list or cust_id in cust_list:
        room_no = random.randrange(60)+300
        cust_id = random.randrange(60)+10

    print("")
    print("******ROOM BOOKED SUCCESSFULLY******\n")
    print("Your assigned room: ",room_no)
    print("Your assigned ID: ",cust_id)
    room_list.append(room_no)
    cust_list.append(cust_id)
```
